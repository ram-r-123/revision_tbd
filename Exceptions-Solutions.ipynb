### Quadratic equation

Consider the function ```quad()``` below that computes the roots of a quadratic equation. Write a program that asks the user for the parameters, calls the function, prints the results and handles all exceptions. In case of error, your program can extract the offending parameters from the ```.params``` attribute of the exception and, for instance, print them (note that since ```a```, ```b```, and ```c``` are not passed to the constructor of ```QuadError``` but stored in an attribute, they do not get printed automatically when the exception object is printed).

````python
import math

class QuadError( Exception ): 
    pass # do nothing special

def quad(a,b,c):
    """ Solve the quadratic equation ax^2 + bx + c=0 """
    if a == 0:
        ex= QuadError( "Not Quadratic" )
        ex.params= ( a, b, c ) # may help with debugging
        raise ex
    if b*b-4*a*c < 0:
        ex= QuadError( "No Real Roots" )
        ex.params= ( a, b, c )
        raise ex
    x1= (-b+math.sqrt(b*b-4*a*c))/(2*a)
    x2= (-b-math.sqrt(b*b-4*a*c))/(2*a)
    return (x1,x2)
````

````python
def input_float(message):
    while True:
        try:
            strinput=input(message)
            return float(strinput) # obviously exits the loop
        except ValueError:
            print("please be so kind as to enter a number")
    
print ("Please enter parameters a, b, c for ax^2 + bx + c = 0 :")
a=input_float("a = ")
b=input_float("b = ")
c=input_float("c = ")

try:
    x1,x2=quad(a, b, c)
    print("The roots of your equation are:")
    print(x1, x2)
except QuadError as e:
    if e.args[0]=='No Real Roots':
        print ("Your equation has no real roots")
    elif e.args[0]=="Not Quadratic":
        print("Please set a!=0 - this equation is not quadratic")
    print("you entered:")
    print(f"a = {a}, b = {b}, c = {c}")
````

# File input/output

A typical situation in which exceptions can occur is when handling files. For instance, the file you want to read may not exist, or you may not have write permissions for the directory to which you are trying to write.

Take some code you have written to read a text file and change the name of the file to something non-existing. See what happens. Modify the code so that the exception is handled correctly (for instance, by asking the user for a new filename or printing a message and quitting cleanly).

````python
# This is just one of many options: handle the exception within the function
# and return an error code that the caller checks for when the function returns. 

def read_fasta(filename):
    try:
        FASTA=open(filename, "r")
        header=FASTA.readline()
        protein="" # build up the sequence here
        for ll in FASTA:
            protein+=ll.rstrip() # remove trailing '\n'
        FASTA.close()
    except OSError as e:
        # OSError is a superclass of some common file errors, see
        # https://docs.python.org/3/library/exceptions.html
        return {"error": e.args}
    # Done. This is just pretty-printing
    (code, name)= header.split('|')
    return {"name": name,
            "accession": code,
            "sequence": protein}


pdata=read_fasta("P04637.fa") # s missing from filename
if not "error" in pdata:
    print("Accession code:")
    print(pdata['accession'])
    print("\nName:")
    print(pdata['name'])
    print("Sequence:")
    print(pdata['sequence'])
else:
    print("An error occurred: ", pdata['error'])
````
